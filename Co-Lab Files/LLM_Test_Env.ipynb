```python
# CLAUDE

import re


class JavaCodeSmellDetector_Claude:
    def __init__(self):
        # Thresholds for detection
        self.COMPLEXITY_THRESHOLD = 10
        self.PARAMETER_THRESHOLD = 5
        self.RESPONSIBILITY_THRESHOLD = 3

    def detect_smells(self, java_code):
        """Main method to detect all code smells in Java code."""
        smells = []

        # Detect each type of smell
        if self.has_complex_method(java_code):
            smells.append("Complex Method")

        if self.has_long_parameter_list(java_code):
            smells.append("Long Parameter List")

        if self.has_multifaceted_abstraction(java_code):
            smells.append("Multifaceted Abstraction")

        # Format output
        return self.format_output(smells)

    def has_complex_method(self, code):
        """Detect if any method has high cyclomatic complexity."""
        methods = self.extract_methods(code)

        for method in methods:
            complexity = self.calculate_cyclomatic_complexity(method)
            if complexity > self.COMPLEXITY_THRESHOLD:
                return True
        return False

    def calculate_cyclomatic_complexity(self, method_code):
        """Calculate cyclomatic complexity of a method."""
        # Start with base complexity of 1
        complexity = 1

        # Decision points that increase complexity
        patterns = [
            r'\bif\s*\(',
            r'\belse\s+if\s*\(',
            r'\bfor\s*\(',
            r'\bwhile\s*\(',
            r'\bcase\s+',
            r'\bcatch\s*\(',
            r'\&\&',
            r'\|\|',
            r'\?',  # Ternary operator
        ]

        for pattern in patterns:
            matches = re.findall(pattern, method_code)
            complexity += len(matches)

        return complexity

    def has_long_parameter_list(self, code):
        """Detect if any method has a long parameter list."""
        methods = self.extract_methods(code)

        for method in methods:
            param_count = self.count_parameters(method)
            if param_count > self.PARAMETER_THRESHOLD:
                return True
        return False

    def count_parameters(self, method_code):
        """Count the number of parameters in a method signature."""
        # Extract method signature (up to the opening brace or semicolon)
        match = re.search(r'\([^)]*\)', method_code)
        if not match:
            return 0

        params_str = match.group(0)[1:-1].strip()  # Remove parentheses

        if not params_str:
            return 0

        # Split by comma and count, handling generics
        params = self.smart_split_parameters(params_str)
        return len(params)

    def smart_split_parameters(self, params_str):
        """Split parameters handling generics like List<String>."""
        params = []
        current_param = ""
        depth = 0

        for char in params_str:
            if char == '<':
                depth += 1
                current_param += char
            elif char == '>':
                depth -= 1
                current_param += char
            elif char == ',' and depth == 0:
                if current_param.strip():
                    params.append(current_param.strip())
                current_param = ""
            else:
                current_param += char

        if current_param.strip():
            params.append(current_param.strip())

        return params

    def has_multifaceted_abstraction(self, code):
        """Detect if a class has multiple responsibilities."""
        # Check if this is a class definition
        if not re.search(r'\bclass\s+\w+', code):
            return False

        # Count distinct responsibilities by analyzing:
        # 1. Number of distinct method prefixes (get, set, calculate, process, etc.)
        # 2. Presence of multiple unrelated concerns

        methods = self.extract_methods(code)
        if len(methods) == 0:
            return False

        # Extract method names
        method_names = []
        for method in methods:
            match = re.search(r'\b(\w+)\s*\([^)]*\)', method)
            if match:
                method_names.append(match.group(1))

        # Identify distinct responsibility indicators
        responsibilities = set()

        responsibility_patterns = {
            'data_access': ['get', 'set', 'fetch', 'retrieve', 'store', 'save'],
            'calculation': ['calculate', 'compute', 'sum', 'total'],
            'validation': ['validate', 'check', 'verify', 'is', 'has'],
            'formatting': ['format', 'print', 'display', 'show', 'render'],
            'processing': ['process', 'handle', 'execute', 'run', 'perform'],
            'conversion': ['convert', 'transform', 'parse', 'to'],
            'notification': ['notify', 'send', 'alert', 'broadcast'],
            'management': ['create', 'delete', 'update', 'manage', 'add', 'remove']
        }

        for method_name in method_names:
            method_lower = method_name.lower()
            for responsibility, keywords in responsibility_patterns.items():
                for keyword in keywords:
                    if method_lower.startswith(keyword):
                        responsibilities.add(responsibility)
                        break

        # Also check for field declarations indicating different concerns
        field_types = self.extract_field_types(code)
        if len(field_types) > 5:  # Many different field types suggest multiple concerns
            responsibilities.add('data_management')

        return len(responsibilities) > self.RESPONSIBILITY_THRESHOLD

    def extract_methods(self, code):
        """Extract all method definitions from Java code."""
        methods = []

        # Pattern to match method signatures (public/private/protected, return type, name, parameters)
        # This handles both complete classes and standalone methods
        pattern = r'(?:public|private|protected|static|\s)+[\w\<\>\[\]]+\s+(\w+)\s*\([^)]*\)\s*(?:throws\s+[\w\,\s]+)?\s*\{'

        matches = re.finditer(pattern, code)

        for match in matches:
            start = match.start()
            # Find the matching closing brace
            end = self.find_matching_brace(code, match.end() - 1)
            if end != -1:
                methods.append(code[start:end + 1])

        return methods

    def find_matching_brace(self, code, start_pos):
        """Find the position of the matching closing brace."""
        depth = 1
        i = start_pos + 1

        while i < len(code) and depth > 0:
            if code[i] == '{':
                depth += 1
            elif code[i] == '}':
                depth -= 1
            i += 1

        return i - 1 if depth == 0 else -1

    def extract_field_types(self, code):
        """Extract field type declarations from a class."""
        # Pattern to match field declarations
        pattern = r'(?:public|private|protected|static|final|\s)+(\w+(?:<[^>]+>)?)\s+\w+\s*[;=]'
        matches = re.findall(pattern, code)
        return set(matches)

    def format_output(self, smells):
        """Format the output according to specifications."""
        if not smells:
            return "This content is not smelly"
        else:
            smell_list = ", ".join(smells)
            return f"This content is smelly with the following smells: |{smell_list}|"


def CLAUDE(java_code_string):
    """Main function to analyze Java code for smells."""
    detector = JavaCodeSmellDetector_Claude()
    result = detector.detect_smells(java_code_string)
    return result


  #print("Test 1:", analyze_code(test1))
  #print("Test 2:", analyze_code(test2))
  #print("Test 3:", analyze_code(test3))
  #print("Test 4:", analyze_code(test4))
```

```python
#GEMINI

import re

import re

def remove_comments_and_strings_GEMINI(code):
    """
    Removes Java comments (// and /* ... */) and string literals
    to ensure analysis is performed only on logic.
    """
    # Remove string literals
    pattern_string = r'"(?:\\.|[^"\\])*"'
    code = re.sub(pattern_string, '', code)

    # Remove single line comments
    pattern_single = r'//.*'
    code = re.sub(pattern_single, '', code)

    # Remove multi-line comments
    pattern_multi = r'/\*.*?\*/'
    code = re.sub(pattern_multi, '', code, flags=re.DOTALL)

    return code

def analyze_complexity_GEMINI(code):
    """
    Detects 'Complex Method' smell.
    Logic: Calculates Cyclomatic Complexity.
    Threshold: > 10 indicates a complex method.
    """
    # Base complexity is 1
    complexity = 1

    # tokens that increment complexity
    branching_tokens = [
        r'\bif\b', r'\bfor\b', r'\bwhile\b', r'\bcase\b',
        r'\bcatch\b', r'\?', r'&&', r'\|\|'
    ]

    for token in branching_tokens:
        complexity += len(re.findall(token, code))

    return complexity > 10

def analyze_parameter_list_GEMINI(code):
    """
    Detects 'Long Parameter List' smell.
    Logic: Finds method signatures and counts parameters.
    Threshold: > 5 parameters.
    """
    # Regex to find method signatures
    method_sig_pattern = r'\b(?:public|private|protected|static|void|int|String|boolean|\w+)\s+\w+\s*\(([^)]*)\)'

    matches = re.findall(method_sig_pattern, code)

    for params in matches:
        if not params.strip():
            continue
        param_count = len(params.split(','))
        if param_count > 5:
            return True

    return False

def analyze_multifaceted_abstraction_GEMINI(code):
    """
    Detects 'Multifaceted Abstraction' smell.
    Logic: Checks if a Class has too many responsibilities.
    Threshold: Combined method and field count > 20.
    """
    # Check if this is a class definition
    if not re.search(r'\bclass\b', code):
        return False

    # Count methods (rough approximation by looking for signatures followed by {)
    method_pattern = r'\b(?:public|private|protected)\s+(?:[\w<>]+\s+)+\w+\s*\([^)]*\)\s*\{'
    method_count = len(re.findall(method_pattern, code))

    # Count fields (rough approximation: access modifier + type + name + semicolon)
    field_pattern = r'\b(?:public|private|protected)\s+(?:static\s+|final\s+)*[\w<>]+\s+\w+\s*;'
    field_count = len(re.findall(field_pattern, code))

    return (method_count + field_count) > 20

def GEMINI(java_code_string):
    """
    Main driver function to detect smells and format output.
    """
    smells = []

    # 1. Clean the code for analysis
    clean_code = remove_comments_and_strings_GEMINI(java_code_string)

    # 2. Check for Complex Method
    if analyze_complexity_GEMINI(clean_code):
        smells.append("Complex Method")

    # 3. Check for Long Parameter List
    if analyze_parameter_list_GEMINI(clean_code):
        smells.append("Long Parameter List")

    # 4. Check for Multifaceted Abstraction
    if analyze_multifaceted_abstraction_GEMINI(clean_code):
        smells.append("Multifaceted Abstraction")

    # 5. Format Output
    if not smells:
        return "This content is not smelly"
    else:
        smell_str = ", ".join(smells)
        return f"This content is smelly with the following smells: |{smell_str}|"
```

```python
import re

# Thresholds for detecting smells
COMPLEXITY_THRESHOLD = 10
LONG_PARAM_THRESHOLD = 5
MULTI_METHOD_COUNT = 8
MULTI_PREFIX_COUNT = 4


def strip_comments_and_strings_ChatGPT(code: str) -> str:
    pattern = re.compile(
        r'(\"(?:\\.|[^\"\\])*\"'
        r'|\'.*?\''
        r'|//[^\n]*'
        r'|/\*.*?\*/)',
        re.DOTALL
    )

    def replacer(match: re.Match) -> str:
        s = match.group(0)
        if s.startswith("//") or s.startswith("/*"):
            return ""
        return '""'

    return pattern.sub(replacer, code)


def find_matching_brace_ChatGPT(code: str, open_index: int) -> int:
    depth = 0
    for i in range(open_index, len(code)):
        c = code[i]
        if c == '{':
            depth += 1
        elif c == '}':
            depth -= 1
            if depth == 0:
                return i
    return len(code) - 1


METHOD_WITH_RET_PATTERN = re.compile(
    r'''
    ^\s*
    (?:@[A-Za-z0-9_\.]+ \s*)*
    (?:public|protected|private|static|final|
       abstract|synchronized|native|transient|
       strictfp)?\s*
    (?:<[^>]+>\s*)?
    [A-Za-z_][\w\<\>\[\]]*
    \s+
    (?P<name>[A-Za-z_]\w*)
    \s*
    \(
        (?P<params>[^)]*)
    \)
    [^{;]*\{
    ''',
    re.MULTILINE | re.VERBOSE
)

CONSTRUCTOR_PATTERN = re.compile(
    r'''
    ^\s*
    (?:@[A-Za-z0-9_\.]+ \s*)*
    (?:public|protected|private)?\s*
    (?P<name>[A-Za-z_]\w*)
    \s*
    \(
        (?P<params>[^)]*)
    \)
    [^{;]*\{
    ''',
    re.MULTILINE | re.VERBOSE
)


def extract_methods_ChatGPT(cleaned_code: str):
    methods = []
    seen = set()

    for pattern in (METHOD_WITH_RET_PATTERN, CONSTRUCTOR_PATTERN):
        for m in pattern.finditer(cleaned_code):
            name = m.group('name')
            params = m.group('params')

            sig_end = m.end()
            open_brace_index = cleaned_code.find('{', sig_end - 1)
            if open_brace_index == -1:
                continue

            key = (name, open_brace_index)
            if key in seen:
                continue
            seen.add(key)

            close_brace_index = find_matching_brace_ChatGPT(cleaned_code, open_brace_index)
            body = cleaned_code[open_brace_index:close_brace_index + 1]
            methods.append({'name': name, 'params': params, 'body': body})

    return methods


def count_params_ChatGPT(param_str: str) -> int:
    param_str = param_str.strip()
    if not param_str:
        return 0
    parts = [p.strip() for p in param_str.split(',')]
    return sum(1 for p in parts if p)


def cyclomatic_complexity_ChatGPT(body: str) -> int:
    complexity = 1
    complexity += len(re.findall(r'\b(if|for|while|case|catch)\b', body))
    complexity += len(re.findall(r'&&|\|\|', body))
    complexity += body.count('?')
    return complexity


def method_prefix_ChatGPT(name: str) -> str:
    prefix_chars = []
    for i, ch in enumerate(name):
        if i > 0 and ch.isupper():
            break
        if ch == '_':
            break
        prefix_chars.append(ch)
    prefix = "".join(prefix_chars)
    return prefix if prefix else name


def detect_multifaceted_abstraction_ChatGPT(original_code: str, methods) -> bool:
    if ('class ' not in original_code
            and 'interface ' not in original_code
            and 'enum ' not in original_code):
        return False

    if len(methods) < MULTI_METHOD_COUNT:
        return False

    prefixes = {method_prefix_ChatGPT(m['name']) for m in methods}
    return len(prefixes) >= MULTI_PREFIX_COUNT


def CHATGPT(java_code: str) -> str:
    cleaned = strip_comments_and_strings_ChatGPT(java_code)
    methods = extract_methods_ChatGPT(cleaned)

    has_complex = False
    has_long_params = False

    for m in methods:
        if cyclomatic_complexity_ChatGPT(m['body']) > COMPLEXITY_THRESHOLD:
            has_complex = True
        if count_params_ChatGPT(m['params']) > LONG_PARAM_THRESHOLD:
            has_long_params = True

    has_multifaceted = detect_multifaceted_abstraction_ChatGPT(java_code, methods)

    smells = []
    if has_complex:
        smells.append("Complex Method")
    if has_long_params:
        smells.append("Long Parameter List")
    if has_multifaceted:
        smells.append("Multifaceted Abstraction")

    if not smells:
        return "This content is not smelly"
    return "This content is smelly with the following smells: |" + ", ".join(smells) + "|"
```

```python
#DEEPSEEK
import re

class JavaCodeSmellDetector_DEEPSEEK:
    def __init__(self):
        # Thresholds for detection
        self.COMPLEXITY_THRESHOLD = 10  # Cyclomatic complexity threshold
        self.PARAMETER_THRESHOLD = 5    # Long parameter list threshold
        self.METHOD_COUNT_THRESHOLD = 8  # For multifaceted abstraction

    def detect_smells(self, java_code):
        """
        Main function to detect code smells in Java code
        """
        smells = []

        # Clean and prepare the code
        cleaned_code = self._clean_code(java_code)

        # Check if the content is a class or just a method
        is_class = "class " in cleaned_code

        # Detect Complex Method
        if self._has_complex_method(cleaned_code):
            smells.append("Complex Method")

        # Detect Long Parameter List
        if self._has_long_parameter_list(cleaned_code):
            smells.append("Long Parameter List")

        # Detect Multifaceted Abstraction (only for classes)
        if is_class and self._has_multifaceted_abstraction(cleaned_code):
            smells.append("Multifaceted Abstraction")

        # Generate output
        if smells:
            return f"This content is smelly with the following smells: |{', '.join(smells)}|"
        else:
            return "This content is not smelly"

    def _clean_code(self, code):
        """Remove comments and normalize whitespace"""
        # Remove single-line comments
        code = re.sub(r'//.*', '', code)
        # Remove multi-line comments
        code = re.sub(r'/\*.*?\*/', '', code, flags=re.DOTALL)
        # Normalize whitespace
        code = re.sub(r'\s+', ' ', code)
        return code.strip()

    def _has_complex_method(self, code):
        """
        Detect methods with high cyclomatic complexity
        Complexity is approximated by counting decision points
        """
        # Find method declarations
        method_pattern = r'(public|private|protected|static|\s) +[\w<>\[\]]+\s+(\w+)\s*\([^)]*\)\s*\{'
        methods = re.finditer(method_pattern, code)

        for method in methods:
            # Find the method body (from { to matching })
            start_pos = method.end() - 1  # Position of the opening brace
            brace_count = 0
            in_string = False
            string_char = None

            for i in range(start_pos, len(code)):
                char = code[i]

                # Handle string literals
                if char in ['"', "'"] and (i == 0 or code[i-1] != '\\'):
                    if not in_string:
                        in_string = True
                        string_char = char
                    elif char == string_char:
                        in_string = False
                        string_char = None

                if not in_string:
                    if char == '{':
                        brace_count += 1
                    elif char == '}':
                        brace_count -= 1
                        if brace_count == 0:
                            method_body = code[start_pos:i+1]
                            complexity = self._calculate_complexity(method_body)
                            if complexity > self.COMPLEXITY_THRESHOLD:
                                return True
                            break
        return False

    def _calculate_complexity(self, method_body):
        """Calculate cyclomatic complexity by counting decision points"""
        complexity = 1  # Base complexity

        # Count if statements
        complexity += len(re.findall(r'\bif\s*\(', method_body))

        # Count for loops
        complexity += len(re.findall(r'\bfor\s*\(', method_body))

        # Count while loops
        complexity += len(re.findall(r'\bwhile\s*\(', method_body))

        # Count do-while loops
        complexity += len(re.findall(r'\bdo\s*\{', method_body))

        # Count switch cases
        switch_matches = re.findall(r'\bswitch\s*\(', method_body)
        complexity += len(switch_matches)

        # Count catch blocks
        complexity += len(re.findall(r'\bcatch\s*\(', method_body))

        # Count ternary operators
        complexity += len(re.findall(r'\?', method_body)) // 2

        # Count logical operators that create branches
        and_or_count = len(re.findall(r'&&|\|\|', method_body))
        complexity += and_or_count // 2

        return complexity

    def _has_long_parameter_list(self, code):
        """Detect methods with too many parameters"""
        # Pattern to find method declarations with parameters
        method_pattern = r'(public|private|protected|static|\s) +[\w<>\[\]]+\s+(\w+)\s*\(([^)]*)\)'

        for match in re.finditer(method_pattern, code):
            params_str = match.group(3).strip()
            if params_str:  # If there are parameters
                # Split by commas and count
                params = [p.strip() for p in params_str.split(',') if p.strip()]
                # Filter out empty parameters and count
                param_count = len([p for p in params if p and p not in ['final', 'static', 'public', 'private', 'protected']])
                if param_count > self.PARAMETER_THRESHOLD:
                    return True
        return False

    def _has_multifaceted_abstraction(self, code):
        """
        Detect classes with too many responsibilities
        Approximated by counting number of methods and diversity of operations
        """
        # Extract class name
        class_match = re.search(r'class\s+(\w+)', code)
        if not class_match:
            return False

        # Count all methods in the class
        method_pattern = r'(public|private|protected|static|\s) +[\w<>\[\]]+\s+(\w+)\s*\([^)]*\)\s*\{'
        methods = re.findall(method_pattern, code)

        # Count getters and setters
        getter_setter_pattern = r'(get|set|is)[A-Z]\w*\s*\('
        getter_setters = re.findall(getter_setter_pattern, code, re.IGNORECASE)

        # Count other methods (excluding getters/setters)
        other_methods = len(methods) - len(getter_setters)

        # Heuristic: If there are many methods that aren't simple getters/setters,
        # and there's a mix of different operation types, it might be multifaceted
        if other_methods > self.METHOD_COUNT_THRESHOLD:
            # Check for diverse method names (crude approximation of multiple responsibilities)
            method_names = [m[1] for m in methods]  # m[1] is the method name group
            unique_prefixes = set()

            for name in method_names:
                # Extract prefix/type of method (first word-like pattern before uppercase)
                if name.lower().startswith(('get', 'set', 'is')):
                    unique_prefixes.add('accessor')
                elif name.lower().startswith(('create', 'make', 'build')):
                    unique_prefixes.add('creator')
                elif name.lower().startswith(('save', 'load', 'store', 'read', 'write')):
                    unique_prefixes.add('persistence')
                elif name.lower().startswith(('validate', 'check', 'verify')):
                    unique_prefixes.add('validator')
                elif name.lower().startswith(('calculate', 'compute', 'process')):
                    unique_prefixes.add('processor')
                else:
                    unique_prefixes.add('other')

            # If we have many different types of operations, it might be doing too much
            if len(unique_prefixes) >= 4:  # Arbitrary threshold
                return True

        return False


# Main function to use the detector
def DEEPSEEK(java_code_string):
    """
    Analyze Java code for code smells

    Args:
        java_code_string (str): Java code as a string (can be class or method)

    Returns:
        str: Detection result in the specified format
    """
    detector = JavaCodeSmellDetector_DEEPSEEK()
    return detector.detect_smells(java_code_string)

```

```python
#GROK
import re
from collections import defaultdict

def extract_class_body_grok(code_str):
    match = re.search(r'(?P<modifiers>(?:(?:public|private|protected|abstract|final|strictfp)\s+)*)class\s+(?P<name>\w+)\s*(?:extends\s+[\w<>\[\]]+)?\s*(?:implements\s+[\w<>\[\]]+(?:\s*,\s*[\w<>\[\]]+)*)?\s*{', code_str)
    if not match:
        return None
    start = match.end()
    brace_count = 1
    i = start
    while i < len(code_str):
        if code_str[i] == '{':
            brace_count += 1
        elif code_str[i] == '}':
            brace_count -= 1
            if brace_count == 0:
                return code_str[start:i]
        i += 1
    return None

def extract_methods_grok(code_body):
    methods = []
    pos = 0
    while True:
        match = re.search(r'(?P<modifiers>(?:(?:public|private|protected|static|final|synchronized|abstract|native)\s+)*)(?P<return>[\w<>\[\]]+)\s+(?P<name>\w+)\s*\((?P<params>[^)]*)\)\s*(?P<throws>throws\s+[\w<>\[\]]+(?:\s*,\s*[\w<>\[\]]+)*)?\s*{', code_body[pos:])
        if not match:
            break
        rel_start = match.end()
        start = pos + rel_start - 1
        brace_count = 1
        i = start + 1
        while i < len(code_body):
            if code_body[i] == '{':
                brace_count += 1
            elif code_body[i] == '}':
                brace_count -= 1
                if brace_count == 0:
                    body = code_body[start:i+1]
                    methods.append({
                        'name': match.group('name'),
                        'params': match.group('params'),
                        'body': body
                    })
                    pos += i + 1
                    break
            i += 1
        else:
            break
    return methods

def extract_fields_grok(class_body):
    field_pattern = r'(?P<modifiers>(?:(?:public|private|protected|static|final)\s+)*)(?P<type>[\w<>\[\]]+)\s+(?P<name>\w+)\s*(?:= [^;]*)?;'
    return [m.group('name') for m in re.finditer(field_pattern, class_body) if 'static' not in m.group('modifiers')]

def calculate_cyclomatic_grok(body):
    if not body:
        return 1
    control_patterns = [
        r'\bif\b', r'\bwhile\b', r'\bdo\b', r'\bfor\b', r'\bswitch\b',
        r'\bcase\b', r'\bcatch\b', r'\?\s*:',
        r'&&', r'\|\|'
    ]
    complexity = 1
    for pattern in control_patterns:
        complexity += len(re.findall(pattern, body))
    switch_cases = len(re.findall(r'\bcase\b', body))
    if switch_cases > 0:
        complexity += switch_cases - len(re.findall(r'\bswitch\b', body))
    return complexity

def calculate_lcom_grok(fields, methods):
    if not fields or len(methods) < 2:
        return 1
    field_set = set(fields)
    method_fields = {}
    for method in methods:
        if not method['body']:
            continue
        accessed = set()
        for field in field_set:
            if re.search(r'(?<!\w)(this\.)?' + re.escape(field) + r'(?!\w)', method['body']):
                accessed.add(field)
        method_fields[method['name']] = accessed
    if not method_fields:
        return 1
    graph = defaultdict(list)
    method_names = list(method_fields.keys())
    for i in range(len(method_names)):
        for j in range(i + 1, len(method_names)):
            if method_fields[method_names[i]] & method_fields[method_names[j]]:
                graph[method_names[i]].append(method_names[j])
                graph[method_names[j]].append(method_names[i])
    visited = set()
    components = 0
    for m in method_names:
        if m not in visited:
            components += 1
            stack = [m]
            while stack:
                curr = stack.pop()
                if curr not in visited:
                    visited.add(curr)
                    for neigh in graph[curr]:
                        if neigh not in visited:
                            stack.append(neigh)
    return components

def GROK(code_str):
    smells = set()
    class_body = extract_class_body_grok(code_str)
    is_method_only = class_body is None

    if is_method_only:
        methods = extract_methods_grok(code_str)
        if methods:
            method = methods[0]
            param_list = [p.strip() for p in method['params'].split(',') if p.strip()]
            if len(param_list) > 4:
                smells.add("Long Parameter List")
            if calculate_cyclomatic_grok(method['body']) > 10:
                smells.add("Complex Method")
    else:
        fields = extract_fields_grok(class_body)
        methods = extract_methods_grok(class_body)
        if calculate_lcom_grok(fields, methods) > 1:
            smells.add("Multifaceted Abstraction")
        for method in methods:
            param_list = [p.strip() for p in method['params'].split(',') if p.strip()]
            if len(param_list) > 4:
                smells.add("Long Parameter List")
            if calculate_cyclomatic_grok(method['body']) > 10:
                smells.add("Complex Method")

    if not smells:
        return "This content is not smelly"
    else:
        smells_list = ", ".join(sorted(smells))
        return f"This content is smelly with the following smells: |{smells_list}|"
```

```python
def TestLLMs(java_code):
    results = {}

    results["GEMINI"] = GEMINI(java_code)
    results["CHATGPT"] = CHATGPT(java_code)
    results["DEEPSEEK"] = DEEPSEEK(java_code)
    results["CLAUDE"] = CLAUDE(java_code)
    results["GROK"] = GROK(java_code)

    return results
```
