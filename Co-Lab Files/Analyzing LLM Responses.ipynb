```python
class ConfusionMatrix:
    def __init__(self):
        self.TP = 0
        self.FP = 0
        self.FN = 0
        self.TN = 0

    def add_TP(self):
        self.TP += 1

    def add_FP(self):
        self.FP += 1

    def add_FN(self):
        self.FN += 1

    def add_TN(self):
        self.TN += 1

    def precision(self):
        return self.TP / (self.TP + self.FP) if (self.TP + self.FP) > 0 else 0

    def recall(self):
        return self.TP / (self.TP + self.FN) if (self.TP + self.FN) > 0 else 0

    def f1(self):
        p = self.precision()
        r = self.recall()
        return 2 * (p * r) / (p + r) if (p + r) > 0 else 0
```
